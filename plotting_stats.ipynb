```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOADING DATA

```python
df = pd.read_csv("cleaned_data.csv")
df
```

# CHECKING DATA

```python
rows = df[df['region'] == 'Астана']
print(len(rows))
rows
```

```python
# how data is distributed?

region_counts = df['region'].value_counts()
total_rows = len(df)

for value, count in region_counts.items():
    percentage = (count / total_rows) * 100
    print(f"{percentage:.1f}% of values in 'region' column are {value}, {count}")
```

# PLOTTING STUFF

```python
plt.rcParams['figure.dpi'] = 300

# colors = sns.color_palette("YlOrRd", len(df['region'].unique()))

colors = sns.color_palette("muted")
```

```python
plt.figure(figsize=(10, 8))

value_counts = df['region'].value_counts()
percentages = value_counts / len(df) * 100

filtered_regions = percentages[percentages >= 4].index

plt.pie(value_counts, labels=None, autopct=lambda p: '{:.1f}%'.format(p) if p >= 3.9 else '', colors=colors, startangle=140)

plt.legend(df['region'].unique(), bbox_to_anchor=(1, 0.5), loc="center left", title="Regions")

plt.title('Distribution of Regions')

plt.show()
```
